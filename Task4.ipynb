# Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    # Rule 1: Amount must be ≥ 50
    if amount < 50:
        return False
    # Rule 2: Should be a valid plan
    if amount not in valid_plans:
        return False
    return True
# ---- Retry System ----
while True:
    amount = int(input("Enter recharge amount: "))
    if validate_recharge(amount):
        print("Recharge successful!")
        break
    else:
        print("Invalid amount! Try again.\n")
```

# Inventory Reorder Alert System

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")
# Example dictionary
inventory = {
    "Apples": 10,
    "Bananas": 25,
    "Milk": 5,
    "Bread": 20
}
check_inventory(inventory)
```

# Student Result Processing System

```python
def student_result(marks):
    total = 0
    # Using loop to calculate total marks
    for m in marks:
        total += m
    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"
#Input
marks_list = [60, 55, 40, 70]
print("Result:", student_result(marks_list))
```

# Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak_hour):
    base_fare = 50
    per_km = 12
    fare = base_fare + (distance * per_km)
    # Peak hour extra 25%
    if peak_hour == "yes":
        fare = fare + (fare * 0.25)
    return fare
# Retry system
while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    final_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", final_fare)
    again = input("Calculate again? (yes/no): ").lower()
    if again != "yes":
        break
```

# Employee Attendance Eligibility Checker


```python
def attendance_checker(attendance_list):
    present_count = 0
    # Count number of P using loop
    for day in attendance_list:
        if day == "P":
            present_count += 1
    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
# Example
attendance = ["P", "A", "P", "P", "P", "A", "P", "P"]
print("Result:", attendance_checker(attendance))
```

# Password Strength Checker

```python
def password_strength(password):
    has_digit = False
    has_special = False
    specials = "@#$"
    # Rule 1: Length check
    if len(password) < 8:
        return "Weak Password"
    # Check each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in specials:
            has_special = True
    # Final check
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
# Example
pwd = input("Enter password: ")
print(password_strength(pwd))
```
