# Social Media – Post Engagement Analyzer

```python
def post_analyzer(likes):
    like_count=0
    #count the likes
    for like in likes:
        like_count+=like
    #check for reel status
    if like_count >=1000:
        return like_count,"Viral post"
    else:
        return like_count,"Normal engagement"
likes=[100,200,600]
total_likes,result=post_analyzer(likes)
print("Likes=",total_likes)
print(result)
```

# Agriculture – Rainfall Adequacy Checker

```python
def rain_checker(rainfall):
    required_level=65
    total_fall=0
    
    for rain in rainfall:
        total_fall+=rain
    avg_rainfall=total_fall/len(rainfall)
    if avg_rainfall >= required_level:
        status="Adequate rainfall"
    else:
        status="Inadequate rainfall"
    return total_fall,avg_rainfall,status
rain_fall=[]
days=int(input("Enter the no. of days:"))
for i in range(days):
    rainfall_num=int(input(f"Enter rainfall for day{i+1}(kg):"))
    rain_fall.append(rainfall_num)
total_fall, avg_rainfall,result=rain_checker(rain_fall)
print("Average rainfal:",avg_rainfall)
print("Rainfall status:",result)
```

# Social Media – Duplicate Account Detection

```python
def detection(username):
    unique_users=set(username)
    if len(username)!=len(unique_users):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
usernames=["alice255","bob1234","alex41"]
detection(usernames)
```

# Healthcare – Appointment Eligibility Checker

```python
def appointment_eligibility(age,emergency_status):
    if age >=18:
        if emergency_status=="emergency":
            status="Appointment Approved"
        elif emergency_status=="not emergency":
            status="Appointment on hold"
        else:
            status="Appoitnment cancelled"
    elif age<18:
        status="Appointment cancelled"
    else:
        return "Invalid age"
    return status
age=int(input("Enter age:"))
emergency_status=input("Enter the emergency status(emergency/not emergency):").lower()
result=appointment_eligibility(age,emergency_status)
print(result)
```

# Agriculture – Premium Crop Price Filter

```python
def crop_price_filter(prices):
    premium=[]
    for price in prices:
        if price>2000:
            premium.append(price)
    return premium
crop_prices=[2400,5000,700,140]
print("Premium crops:",crop_price_filter(crop_prices))
```

# System Monitoring – Application Health Checker

# Banking – Daily Transaction Limit Checker

# Smart Electricity Bill Analyzer

```python
def bill_analyzer(units):
    unit_count=sum(units)
    #calculating the units
    if unit_count<=100:
        total_bill=unit_count*3
    elif unit_count>100 and unit_count<=200:
        total_bill=(100*3)+((unit_count-100)*5)
    elif unit_count>200:
        total_bill=(100*3)+(100*5)+((unit_count-200)*7)
    else:
        return "Invalid units"
    #usage status
    if total_bill<500:
        status="Low usage"
    elif total_bill>500 and total_bill<1500:
        status="Moderate usage"
    else:
        status="High usage"
    return total_bill,status
units=[]
days=int(input("Enter the no. of days:"))
for i in range(days):
    units_count=int(input(f"Enter the unit on day {i+1}:"))
    units.append(units_count)
total_bill,result=bill_analyzer(units)
print("Total bill:",total_bill)
print("Usage:",result)
```
