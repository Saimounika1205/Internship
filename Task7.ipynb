#  Smart Light Controller

```python
class SmartLight:
    def __init__(self,light_name):
        self.light_name=light_name
        self.status="OFF" #default
    def turn_on(self):
        self.status="ON"
    def turn_off(self):
        self.status="OFF"
    def display(self):
        print(self.light_name,"is",self.status)
light_name="Bedroom Light"
status="ON"
light=SmartLight(light_name)
if status=="ON":
    light.turn_on()
elif status=="OFF":
    light.turn_off()
else:
    print("Invalid status")
light.display()
    
```

#  Employee ID Card System

```python
class EmployeeCard:
    def __init__(self,name,id,department):
        self.name=name
        self.id=id
        self.department=department
    def display(self):
        print("Employee ID Card")
        print("Name:",self.name)
        print("ID:",self.id)
        print("Department:",self.department)
employee_id=EmployeeCard("Rahul","EMP102","IT")
employee_id.display()
```

# Mobile Contact Record

```python
class Contact:
    def __init__(self,name,ph_no):
        self.name=name
        self.ph_no=ph_no
    def display(self):
        print("Contact Saved")
        print("Name:",self.name)
        print("Phone:",self.ph_no)
contacts=Contact("Anita",9876543210)
contacts.display()
```

# Product Price Tag Generator

```python
class PriceTag:
    def __init__(self,product_name,price):
        self.product_name=product_name
        self.price=price
    def display(self):
        print("Product:",self.product_name)
        print("Price:","₹",self.price)
pricetags=PriceTag("Headphones",2499)
pricetags.display()
```

# Movie Rating Display System

```python
class MovieRating:
    def __init__(self,movie_name,rating):
        self.movie_name=movie_name
        self.rating=rating
    def display(self):
        print("Movie:",self.movie_name)
        print("Rating:",self.rating,"/5")
movie_rating=MovieRating("Inception",4.8)
movie_rating.display()
```

# Delivery Address Manager

```python
class DeliveryAddress:
    def __init__(self,customer_name,address):
        self.customer_name=customer_name
        self.address=address
    def display(self):
        print("Delivery Details")
        print("Customer:",self.customer_name)
        print("Address:",self.address)
delivery=DeliveryAddress("Suman","Hyderabad")
delivery.display()
```
