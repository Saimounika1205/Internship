# Smart Parking Lot Management System

```python
def parking_management(capacity,vehicle_logs):
    parked=0
    for log in vehicle_logs:
        if log=="IN":
            parked+=1
        elif log=="OUT":
            if parked>0:
                parked-=1
            else:
                return "No vehicles"

    #check parking lot status
    if parked>capacity:
        status="Capacity exceeded"
    elif parked==capacity:
        status="Full"
    else:
        status="Available"
    return parked,status
capacity=50
vehicle_logs=["IN","IN","IN","OUT","IN","IN","OUT"]
parked,result=parking_management(capacity,vehicle_logs)
print("Currently Parked Vehicles:",parked)
print("Parking Status:",result)
```

# Online Food Delivery Time Estimator

```python
def delivery_time_estimator(distance,traffic_level,weather):
    time=distance*5
    total_time=time

    #traffic constraints
    if traffic_level=="low":#add 5 minutes
        total_time+=5
    elif traffic_level=="medium":#add 8 minutes
        total_time+=8
    elif traffic_level=="high":#add 10 minutes
        total_time+=10
        
    #weather constraints
    if weather=="normal":
        total_time=time
    elif weather=="rainy": #add 5 minutes
        total_time+=5
    elif weather=="heavy rain": #add 15minutes
        total_time+=15
    #return total
    return total_time
distance=8
traffic_level='high'
weather='rainy'
result=delivery_time_estimator(distance,traffic_level,weather)
print("Estimated Delivery Time:",result)
```

# Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy_analyzer(total_seats,booked_count):
    booked_seats=[1]*booked_count
    occupancy=(len(booked_seats)/total_seats)*100
    if occupancy==100:
        status="Housefull"
        print("Open additional show")
    elif occupancy>=75:
        status="Almost Full"
    else:
        status="Seats available"
    return occupancy,status
total_seats=200
booked_count=150
occupancy,result=seat_occupancy_analyzer(total_seats,booked_count)
print("Occupancy:",int(occupancy),"%")
print("Show Status:",result)
```

# Cloud Server Load Classification System

```python
def load_classification(server_load):
    cpu_load=0
    for load in server_load:
        cpu_load+=load
    avg_cpu=int(cpu_load/len(server_load))
    if avg_cpu<50:
        status="Normal"
    elif avg_cpu>=50 and avg_cpu<=80:
        status="Warning"
    elif avg_cpu==80:
        status="Critical"
    return avg_cpu,status
server_load=[45,60,70,85,90]
average,result=load_classification(server_load)
print("Average CPU Load:",average,"%")
print("Server Status:",result)
```

# Smart Classroom Resource Usage Monitor

```python
def resource_usage(resources):
    thresholds={
        "Projector":6,
        "AC":8,
        "Lights":6
    }
    overuse=[]
    for resource,hours in resources.items():
        if resource in thresholds and hours> thresholds[resource]:
            overuse.append(resource)

    if len(overuse)>0:
        print("Overused:",overuse)
        print("Energy alert:Yes")
    else:
        print("Overused:None")
        print("Energy alert:No")
    return overuse
resources={
    "Projector":6,
    "AC":9,
    "Lights":4
}
result=resource_usage(resources)
```
